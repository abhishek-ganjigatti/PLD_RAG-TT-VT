```python
from google.colab import files
import fitz  # PyMuPDF

# üì§ Upload PDF
uploaded = files.upload()
pdf_path = next(iter(uploaded))
doc = fitz.open(pdf_path)

# üìÑ Extract text
text = ""
for page in doc:
    text += page.get_text()

print("üìÑ Sample extracted text (first 500 chars):\n")
print(text[:500])
```

```python
import re

# üîç Split text into sentences
sentences = re.split(r'(?<=[.!?])\s+(?=[A-Z])', text)

# üì¶ Create chunks with slight overlap
chunks = []
chunk = ""
max_words = 120
overlap = 50

for sentence in sentences:
    if len((chunk + " " + sentence).split()) <= max_words:
        chunk += " " + sentence
    else:
        chunks.append(chunk.strip())
        # Add overlap
        chunk_words = chunk.split()
        chunk = " ".join(chunk_words[-overlap:]) + " " + sentence

if chunk:
    chunks.append(chunk.strip())

print(f"üìö Total Chunks: {len(chunks)}")
print("üìå Example Chunk:\n", chunks[0][:300])
```

```python
from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
doc_embeddings = encoder.encode(chunks, convert_to_tensor=False)
```

```python
import faiss
import numpy as np

dimension = doc_embeddings[0].shape[0]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))

print(f"üîç Index contains {index.ntotal} chunks.")
```

```python
# üìù User query
query = "fertilizer for mango plant?"  # You can change this

# üåø Define allowed agricultural keywords
allowed_keywords = [
    "agriculture", "plant", "fertilizer", "soil", "irrigation", "crop", "farm",
    "farming", "water", "harvest", "pesticide", "manure", "weather", "yield",
    "horticulture", "seeds", "organic", "tractor", "livestock", "market price",
    "grain", "wheat", "rice", "maize", "sorghum", "drought", "drip irrigation"
]

# üîç Check if the query is related to agriculture
query_lower = query.lower()
if not any(keyword in query_lower for keyword in allowed_keywords):
    print("‚ùå Off-topic query detected.")
    print("ü§ñ Sorry, I can only answer agriculture-related questions.")
    exit()  # Exit early if query is off-topic

# Proceed with embedding
```

```python
query_embedding = encoder.encode([query])

top_k = 3
distances, indices = index.search(np.array(query_embedding), top_k)

retrieved_docs = [chunks[i] for i in indices[0]]  # Optional: filter loose matches

print("\nüßæ Retrieved Chunks:\n")
if not retrieved_docs:
    print("‚ö†Ô∏è No relevant information found in the uploaded document.")
else:
    for i, doc in enumerate(retrieved_docs):
        print(f"{i+1}. {doc}\n")
```

LLM model

```python
from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-large")
if retrieved_docs:
    context = "\n".join(retrieved_docs)
    prompt = f"Answer the question based on the context below.\n\nContext:\n{context}\n\nQuestion: {query}"
else:
    prompt = f"The uploaded document does not contain relevant information. However, please try to answer the following agriculture-related question using your general knowledge.\n\nQuestion: {query}"

response = generator(prompt, max_new_tokens=100)
print("\nüß† Generated Answer:")
print(response[0]['generated_text'])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_id = "HuggingFaceH4/zephyr-7b-alpha"

# Load tokenizer & model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",  # Will auto-use GPU
    low_cpu_mem_usage=True
)

# Create text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

# New section

```python
# Optional: Truncate context if prompt gets too long
max_tokens = 4096
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
if input_ids.shape[1] > max_tokens:
    print(f"‚ö†Ô∏è Truncating input from {input_ids.shape[1]} to {max_tokens} tokens.")
    input_ids = input_ids[:, -max_tokens:]

# Generate the answer
response = generator(
    tokenizer.decode(input_ids[0]),
    max_new_tokens=150,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    repetition_penalty=1.1,
)

# Print just the model's reply
generated_text = response[0]['generated_text'].split("<|assistant|>")[-1].strip()

print("\nüß† Zephyr Answer:")
print(generated_text)
```
